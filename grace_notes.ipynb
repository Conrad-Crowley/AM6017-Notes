## Tensor

A *tensor* in this context is similar to a numpy-array, in that it contains numbers. You know already about two-dimensional numpy arrays, but of course we can also easily think of higher-dimensional arrays. Many calculations, in particular in machine learning deal with large scale calculations using tensors. 

## Flow

The term *flow* refers to the fact that tensorflow arranges tensor calculations in a direct graph. The nodes in this graph are operations, and the edges are associates with tensors. Incoming edges are the inputs for a given operation and outgoing edges are the outputs of an operation. 

Why is all this so complicated? The reason is that this way of defining graphs for calculations scales very well even if the objects involved contain very large amounts of data.  It turns out that computers are actually not very fast, if they need to do each operation sequentially.  The secret to fast computation is to do many operations in *parallel*. However, the fundamental problem with parallelization of numerical calculataion is that certain calculations need to be finished, before the next ones can be started.  By representing this interdependency in a directed graph, the computer can parallelize operations much more efficiently, since all interconnections are known. 

The strength of tensorflow is that once a graph is defined, the session which evaluates nodes on the graph can in principle run on extremely parallel hardware. In particular GPUs (Graphics processing units) which were originally developed for computer game graphics, turn out to be ideally suited for the parallel processing required for machine learning applications. It also allows for the work to be efficiently distributed across large comuting farms. Unfortunately, we do not have powerful graphics cards in our lab computers and they are also not neatly inerconnected, but to learn the basic principles, the hardware is more than sufficient.

## Tensor

The term *tensor* is a mathematical term loosely speaking describing an object with a number of indices. For example in differential geometry you would learn about the [Riemann tensor](https://en.wikipedia.org/wiki/Riemann_curvature_tensor) which for a 3 dimensional space is written as 
\begin{equation}
R^i_{jkl}\qquad\qquad i,j,k,l \in 1,\ldots,3
\end{equation}
Since all indices can run from 1 to 4, this single symbol is a collection of $3^4=81$ numbers. In practice it is a huge advantage to deal with a simple symbol, rather than many separate numbers.  

The Riemann tensor would be a *rank 4* tensor, since it has 4 indices.  More familiar examples would be *matrices* (rank 2) and vectors (rank 1).  Scalars are formally rank 0 tensors.

Why do we need tensors in Machine learning?  The reason is that the data we are dealing with is often naturally organised along certain indices.  Consider for example a collection of pictures, which all have the same size. You could think of them as a tensor object $p_{i,j,k,l}$. Then for example $p_{7,28,17,1}$ could refer to the value of the *red* ($l=1$) channel of the pixel at position $(j,k)=(28,17)$ in picture number $i=7$.  As such $p$ is an efficient handle to deal with this collection of data.

## Back-Propagation

In fitting a neural network, backpropagation computes the gradient of the loss function with respect to the weights of the network for a single input–output example, and does so efficiently, unlike a naive direct computation of the gradient with respect to each weight individually. This efficiency makes it feasible to use gradient methods for training multilayer networks, updating weights to minimize loss; gradient descent, or variants such as stochastic gradient descent, are commonly used. 

The backpropagation algorithm works by computing the gradient of the loss function with respect to each weight by the chain rule, computing the gradient one layer at a time, iterating backward from the last layer to avoid redundant calculations of intermediate terms in the chain rule

- $S$ is some activtion function.
- General idea: define cost/loss/error function $C(W^k, \mathbf{b}^k)$; need to minimize it.
- training data points $(\mathbf{x}_1,\ldots, \mathbf{x}_T)$ with correct classification $(c_1,\ldots,c_T)$ with $c_d\in\{0,1\}$
- compare output of network $o^N(\mathbf{x}_d)$ with $c_d$
- calculate cost function for training data:
\begin{equation}
 C_T(W^k, \mathbf{b}^k) = \frac{1}{T} \sum_{d=1}^T (o^N(\mathbf{x}_d)-c_d)^2
\end{equation}
- want modify weights and biases to minimize cost function $C_T$, but how should we do this? 
- make a "small" change of weights/biases
\begin{equation}
w^k_{i,j} \to w^k_{i,j} + \Delta w^k_{i,j}
\end{equation}
- How to know, which of the $\Delta w^k_{i,j}$ should be big/small?
- Check the influence each weight has on cost function:
\begin{equation}
\frac{\partial C_T}{\partial w^k_{i,j}} = \frac{2}{T} \sum_{d=1}^T (o^N(\mathbf{x}_d)-c_d)\frac{\partial o^N(\mathbf{x}_d)}{\partial w^k_{i,j}}
\end{equation}
We know that
\begin{equation}
o^{N} = S\left(\sum w^{N-1}_{1,j} o^{N-1}_j+b^{N}_j\right) 
\end{equation}
and therefore 
\begin{equation}
\frac{\partial o^N(\mathbf{x}_d)}{\partial w^k_{i,j}} =  S'\left(\sum_{j} w^{N-1}_{1,j} 
\left(o^{N-1}_{j'}+b^{N-1}_{j'}\right)\right) \times \begin{cases} 
o^{N-1}_j& \text{if $k=N-1$} \\
\sum_{j'} w^{N-1}_{1,j'} \frac{\partial o^{N-1}_{j'}}{\partial w^k_{i,j}} & \text{else}
\end{cases}
\end{equation}

This means we can calculate $\frac{\partial o^N(\mathbf{x}_d)}{\partial w^k_{i,j}} $ in terms of $\frac{\partial o^{N-1}_{j'}(\mathbf{x}_d)}{\partial w^k_{i,j}}$. We can then redo this operation and express $\frac{\partial o^{N-1}_{j'}(\mathbf{x}_d)}{\partial w^k_{i,j}} $ in terms of $\frac{\partial o^{N-2}_{j'}(\mathbf{x}_d)}{\partial w^k_{i,j}}$, etc, until we reach $k$. 


Most prominent advantages of Backpropagation are:

   Backpropagation is fast, simple and easy to program. It has no parameters to tune apart from the numbers of input. It is a flexible method as it does not require prior knowledge about the network. It is a standard method that generally works well. It does not need any special mention of the features of the function to be learned.

## Overfitting

Overfitting is the production of an analysis that corresponds too closely or exactly to a particular set of data, and may therefore fail to fit to additional data or predict future observations reliably.

In Overfitting, the model tries to learn too many details in the training data along with the noise from the training data. As a result, the model performance is very poor on unseen or test datasets. Therefore, the network fails to generalize the features or patterns present in the training dataset.

The possible reasons for Overfitting in neural networks are as follows:
### 1
The size of the training dataset is small
When the network tries to learn from a small dataset it will tend to have greater control over the dataset & will make sure to satisfy all the data points exactly. So, the network is trying to memorize every single data point and failing to capture the general trend from the training dataset.
The model tries to make predictions on Noisy Data
### 2 
Overfitting also occurs when the model tries to make predictions on data that is very noisy, which is caused due to an overly complex model having too many parameters. So, due to this, the overfitted model is inaccurate as the trend does not reflect the reality present in the data.

Deep neural networks are prone to overfitting because they learn millions or billions of parameters while building the model. A model having this many parameters can overfit the training data because it has sufficient capacity to do so.

The basic idea to deal with the problem of overfitting is to decrease the complexity of the model. To do so, we can make the network smaller by simply removing the layers or reducing the number of neurons.

Curse of dimensionality: In machine learning problems that involve learning a "state-of-nature" from a finite number of data samples in a high-dimensional feature space with each feature having a range of possible values, typically an enormous amount of training data is required to ensure that there are several samples with each combination of values. 

## One-Hot encoding

One-hot encoding is a frequently used method to deal with categorical data. Because many machine learning models need their input variables to be numeric, categorical variables need to be transformed in the pre-processing part.

One-Hot encoding is a form of encoding that takes an ordinal coding system where the $n$ labels are assigned a unique alue $i$ between $0$ to $n-1$ and instead encodes each label as a vector of length $n$ : it has a low value $0$ in all positions except for the $i$th postion which is a high value $1$. In this each label now becmes a discrete probability distribution of possible labels. Therefore can be transferred nicely for comparison with a probability distribution output of a NN.

```python
# python implementation of FFN 
# Layer 1 has weights matrix W1 and bias b1
# Activation fn is RELU

# Layer 2 has weights matrix W2 and bias b2
# Activation fn SOFTMAX

import numpy as np
import tensorflow as tf

def relu(x):
    for i in range(len(x)):
        if x[i] < 0:
            x[i] =0
    return x

def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=0)

def two_layer(x,W1,b1,W2,b2):
    # A_2(W_2(A_1(W_1(x)+b_1))+b_2)
    temp = x
    temp = np.matmul(W1, temp) + b1
    
    temp = np.array(temp)[0]
    # temp is 1xn array now
    temp = relu(temp)
    temp = np.matmul(W2, temp) + b2
    
    temp = np.array(temp)[0]
    # temp is 1xn array now
    
    temp = softmax(temp)
    return temp
```

## Softmax 

Instead of using sigmoid, we will use the Softmax activation function in the output layer in the above example. The Softmax activation function calculates the relative probabilities. 

$\sigma (x)_i= \frac{e^{z_i}}{\sum_{j=1}^n e^{z_j}}$

Note that, when the number of classes is two, it becomes the same as the sigmoid activation function.

The Softmax regression is a form of logistic regression that normalizes an input value into a vector of values that follows a probability distribution whose total sums up to 1.

The function is usually used to compute losses that can be expected when training a data set. Known use-cases of softmax regression are in discriminative models such as Cross-Entropy

That is, Softmax assigns decimal probabilities to each class in a multi-class problem. Those decimal probabilities must add up to 1.0. This additional constraint helps training converge more quickly than it otherwise would.

Softmax assumes that each example is a member of exactly one class. 

```python
def softmax_numpy(inputs):
    return np.exp(inputs)/sum(np.exp(np.inputs), axis=0)
```

```python
def softmax_tensorflow(inputs):
    return tf.exp(inputs)/tf.reducesum(tf.exp(inputs), 1) 
```

## Sigmoid 

The sigmoid function is also called a squashing function as its domain is the set of all real numbers, and its range is (0, 1). Hence, if the input to the function is either a very large negative number or a very large positive number,  the output is always between 0 and 1. Therefore, it is especially used for models where we have to predict the probability as an output.

$ \sigma(x) = \frac{1}{1+e^{-x}} $

If we use a linear activation function in a neural network, then this model can only learn linearly separable problems. However, with the addition of just one hidden layer and a sigmoid activation function in the hidden layer, the neural network can easily learn a non-linearly separable problem. Using a non-linear function produces non-linear boundaries and hence, the sigmoid function can be used in neural networks for learning complex decision functions.

The sigmoid function returns values between 0 and 1, and can be used to convert the inputs of the layer to a probability ditribution. The sigmoid function is used as the final activation function in binaray classification problems.

### Why ReLu?

Usually, a matrix in which most entries are 0 is called a sparse matrix and similarly, we desire a property like this in our neural networks where some of the weights are zero. Sparsity results in concise models that often have better predictive power and less overfitting/noise. In a sparse network, it’s more likely that neurons are actually processing meaningful aspects of the problem. 

The activations functions that were used mostly before ReLU such as sigmoid or tanh activation function saturated. This means that large values snap to 1.0 and small values snap to -1 or 0 for tanh and sigmoid respectively. Further, the functions are only really sensitive to changes around their mid-point of their input, such as 0.5 for sigmoid and 0.0 for tanh. This caused them to have a problem called vanishing gradient problem. Let us briefly see what vanishing gradient problem is. 

Neural Networks are trained using the process gradient descent. The gradient descent consists of the backward propagation step which is basically chain rule to get the change in weights in order to reduce the loss after every epoch. It is important to note that the derivatives play an important role in updating of weights. Now when we use activation functions such as sigmoid or tanh, whose derivatives have only decent values from a range of -2 to 2 and are flat elsewhere, the gradient keeps decreasing with the increasing number of layers.

This reduces the value of the gradient for the initial layers and those layers are not able to learn properly. In other words, their gradients tend to vanish because of the depth of the network and the activation shifting the value to zero. This is called the vanishing gradient problem. 

ReLU, on the other hand, does not face this problem as its slope doesn’t plateau, or “saturate,” when the input gets large. Due to this reason models using ReLU activation function converge faster. 

### Negatives

But there are some problems with ReLU activation function such as exploding gradient. The exploding gradient is opposite of vanishing gradient and occurs where large error gradients accumulate and result in very large updates to neural network model weights during training. Due to this, the model is unstable and unable to learn from your training data.

Also, there is a downside for being zero for all negative values and this problem is called “dying ReLU.”A ReLU neuron is “dead” if it’s stuck in the negative side and always outputs 0. Because the slope of ReLU in the negative range is also 0, once a neuron gets negative, it’s unlikely for it to recover. Such neurons are not playing any role in discriminating the input and is essentially useless. Over time you may end up with a large part of your network doing nothing. The dying problem is likely to occur when the learning rate is too high or there is a large negative bias.

## Layers

The next important ingredient for artificial neural networks are the presence of hidden layers.  We have used inputs ($x_1,x_2$) and it is customary to combine them into an *input layer* (although this layer does not actually have any neurons!). The outputs of the input layer are thus simply the external inputs of the system. The neurons corresponding to the actual readout of the network are combined into an output layer. For a typical binary classifier, the output layer will only consist of a single neuron. In-between the input and output layer, there may be a number of hidden layers. The output of the neurons of one layer feeds into the input of the  next layer, until the output layer is reached. To make this a bit more formal, let us introduce some notation:

| symbol|meaning|
|----| :--- |
| $N$   | number of layers -1|
| $l^k$ | # of nodes in layer $k$|
| $x_i$ | ext. input $1\leq i \leq l^0$|
| $n_i^k$ | node # $i$ in layer $k$|
| $i_{i}^k$ | input for node $n_i^k$ 
| $o_i^k$ | output for node $n_i^k$|
| $b_i^k$ | bias for  node $n_i^k$|
| $w^k_{i,j}$ | weight from $n_j^{k-1}$ to $n_i^{k}$|
| $W^k$ | $l_{k}\times l_{k-1}$ weight matrix

Based on this notation, we can then calculate

\begin{align}
o^0_i &= x_i & i=1,\ldots,l^0\\
o_i^{k} &= S^{k}\left(\sum_k^{l^{k-1}} w^k_{i,j} o^{k-1}_j+ b^k_i\right) &k = 1,\ldots,N; i=1,\ldots,l^k 
\end{align}
or in a more elegant and slightly more general vector notation
\begin{align}
\mathbf{o}^0 &= \mathbf{x} \\
\mathbf{o}^{k} &= \mathbf{S}^{k}\left(W^k \mathbf{o}^{k-1} + \mathbf{b}^{k}\right) &k = 1,\ldots,N;
\end{align}

Here the function $S^{k}(x)$ (respectively $\mathbf{S}^k(\mathbf{v}))$ is the neuron function of the layer $k$.  Let us discuss different options for this function:

whiteboard explaining:

- state space $S$, set of possible states
- at time $t$ we are in a state $s_t\in S$
- action space $A$, set of possible actions that can be taken 
- $a_t \in A$ is the action taken at time $t$
- reward $r_t$ at time 

Return $R$ is either the simple sum of rewards 
$$
R = \sum_{t=0}^\infty  r_t
$$


or more commonly discounted sum of rewards:
$$
R = \sum_{t=0}^\infty \gamma^t r_t
$$
where $0\leq \gamma \leq 1$ is a discount factor. Award in the far future is then weighted less than immediate reward

## Policy

The aim is to find a *policy*
$$
\pi:\; A \times S \to [0,1]
$$

For a given policy, the return $R$ becomes a random variable and we can study the expected return 

$$
V_\pi = \langle R \rangle_\pi = E_\pi(R)
$$

in general, you would like to find a policy, which maximizes the return. 

## Q-learning

Find policy by considering a $Q^\pi(s,a)$ function.  If we are currently in state $s$ then take action $a$ in the next step and after that follow the policy $\pi$, then $Q^\pi(s,a)$ gives the the discounted expected reward $R$.

# The [Reverb](https://deepmind.com/research/open-source/Reverb) replay buffer

In RL we can now create lots of *experience* data, by trying out different steps at a given time, and simply observe what happens as a result of our exploration.  The data generated in this way is in general massive, and it is not clear how we can organize it efficiently. 

In general we have two entities:

- Actors which produce experience data
- Learners which consume experience data

Between that two, we require some infrastructure, which holds the experience data temporarily, and discards irrelevant ones as appropriate.  `Reverb` [(introduced in 2021)](https://arxiv.org/abs/2102.04736) provides this functionality by providing a server to communicate between the two. Note: At the moment (2022), `reverb` only runs on Linux.

In our context, the experience data is organized in *trajectories* of how the system evolves as we push it to the left or to the right (see `agent.collect_data_spec`).  As the agent starts to improve its policy, it will require different data from when it was beginning to explore.  

Analogy: Think of the replay buffer as the set of notes/books you carry around in your bag.  As you learn, this set likely changes, and you will likely today not carry around the same notes/books you had in primary school.
